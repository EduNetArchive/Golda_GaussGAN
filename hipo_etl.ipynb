```python
import os
import numpy as np
import pandas as pd
import hipopy.hipopy as hippy
from time import time

path = '/Users/andrey.golda/Documents/Study/NN/job_5067/output/'
```

Be careful. The hipopy library has a lot of bugs (which does not diminish the achievements of its developer, thank him very much). It is impossible to optimize the script under these conditions. Everything is built on crutches.

```python
def evaluate_one_file(filename):
    counter = 0
    start = time()
    file = hippy.open(f'{path}{filename}/dst.hipo', mode="r")
    file.readBank('REC::Particle')
    
    print('---'*10)
    print(f'\nFile {filename}')
    
    file_events = []
    
    for event in file:
        if counter>1:
            file_events.append(np.array([event['MC::Particle_px'],
                                         event['MC::Particle_py'],
                                         event['MC::Particle_pz']]).reshape(3,4).transpose().flatten())
        counter+=1
        
    print(f'Time for file: {time()-start} sec')
    
    np.array(file_events).dump(f'{path}{filename}/dst.txt')
    
    return file_events
```

```python
all_files_dir = os.listdir(path = path)

for filename in all_files_dir[52:]:
    evaluate_one_file(filename)
```

```python
global_file_events = []
for filename in all_files_dir:
    try:
        file_events = np.load(f'{path}{filename}/dst.txt', 
                                allow_pickle=True)
        global_file_events.append(file_events)
    except:
        None
        
global_file_events = np.array(global_file_events)
global_file_events = global_file_events.reshape((-1, 12))
```

```python
df = pd.DataFrame(global_file_events,
                    columns=['p_x_e',
                             'p_y_e',
                             'p_z_e',
                             'p_x_g1',
                             'p_y_g1',
                             'p_z_g1',
                             'p_x_g2',
                             'p_y_g2',
                             'p_z_g2',
                             'p_x_p',
                             'p_y_p',
                             'p_z_p',])
df
```

```python
df.to_csv(f'{path}3M_epgg.csv', index=False)
```
